```python
"""
FITXA INTERACTIVA D'ESTAD√çSTICA - MATEM√ÄTIQUES BATXILLERAT 2
Sistema d'Aprenentatge Guiat per a Problemes de Probabilitat
============================================================
"""

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.patches import FancyBboxPatch, Rectangle
import random
import time

# Configuraci√≥ d'estil per a millor visualitzaci√≥
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12
plt.rcParams['font.family'] = 'sans-serif'

# Colors per a visualitzaci√≥
COLORS = {
    'principal': '#2E86AB',
    'secundari': '#A23B72',
    'exit': '#2A9D8F',
    'error': '#E63946',
    'neutre': '#F1FAEE',
    'groc': '#F77F00',
    'verd_clar': '#90EE90',
    'vermell_clar': '#FFB6C1'
}

class ProblemaEstadistica:
    def __init__(self, enunciat, tipus, dades, passos_solucio):
        self.enunciat = enunciat
        self.tipus = tipus
        self.dades = dades
        self.passos_solucio = passos_solucio

# BANC DE PROBLEMES PER TIPUS
problemes_bayes = [
    ProblemaEstadistica(
        enunciat="""
üè• PROBLEMA M√àDIC

Una nova prova per detectar una malaltia t√© les seg√ºents caracter√≠stiques:
‚Ä¢ La malaltia afecta l'1,5% de la poblaci√≥
‚Ä¢ Si una persona t√© la malaltia, la prova d√≥na positiu en el 98% dels casos
‚Ä¢ Si una persona NO t√© la malaltia, la prova d√≥na positiu en el 2% dels casos

La Maria es fa la prova i el resultat √©s POSITIU.
‚ùì Quina √©s la probabilitat que la Maria tingui realment la malaltia?
        """,
        tipus="bayes",
        dades={
            'p_malaltia': 0.015,
            'p_positiu_donat_malalt': 0.98,
            'p_positiu_donat_sa': 0.02,
            'nom_condicio': 'Malaltia',
            'nom_test': 'Test Positiu'
        },
        passos_solucio=[
            "1Ô∏è‚É£ Identifiquem les dades:\n   ‚Ä¢ P(Malaltia) = 0,015 = 1,5%\n   ‚Ä¢ P(Positiu|Malaltia) = 0,98 = 98%\n   ‚Ä¢ P(Positiu|No Malaltia) = 0,02 = 2%",
            "2Ô∏è‚É£ Calculem P(Positiu) usant probabilitat total:\n   P(Positiu) = P(Positiu|Malaltia)√óP(Malaltia) + P(Positiu|No Malaltia)√óP(No Malaltia)\n   P(Positiu) = 0,98√ó0,015 + 0,02√ó0,985 = 0,0147 + 0,0197 = 0,0344",
            "3Ô∏è‚É£ Apliquem el teorema de Bayes:\n   P(Malaltia|Positiu) = P(Positiu|Malaltia)√óP(Malaltia) / P(Positiu)\n   P(Malaltia|Positiu) = 0,98√ó0,015 / 0,0344 = 0,0147 / 0,0344 = 0,4273",
            "‚úÖ RESPOSTA: La Maria t√© un 42,73% de probabilitat de tenir la malaltia"
        ]
    ),
    ProblemaEstadistica(
        enunciat="""
üöó PROBLEMA D'ASSEGURANCES

Una asseguradora classifica els seus conductors:
‚Ä¢ Risc baix: 60% dels conductors, 5% tenen accidents a l'any
‚Ä¢ Risc mitj√†: 30% dels conductors, 15% tenen accidents a l'any
‚Ä¢ Risc alt: 10% dels conductors, 40% tenen accidents a l'any

Un conductor ha tingut un accident aquest any.
‚ùì Quina √©s la probabilitat que sigui un conductor de risc alt?
        """,
        tipus="bayes",
        dades={
            'grups': ['Baix', 'Mitj√†', 'Alt'],
            'p_grup': [0.60, 0.30, 0.10],
            'p_accident': [0.05, 0.15, 0.40],
            'nom_condicio': 'Risc Alt',
            'nom_test': 'Accident'
        },
        passos_solucio=[
            "1Ô∏è‚É£ Identifiquem les dades:\n   ‚Ä¢ P(Baix) = 0,60 amb P(Accident|Baix) = 0,05\n   ‚Ä¢ P(Mitj√†) = 0,30 amb P(Accident|Mitj√†) = 0,15\n   ‚Ä¢ P(Alt) = 0,10 amb P(Accident|Alt) = 0,40",
            "2Ô∏è‚É£ Calculem P(Accident) per probabilitat total:\n   P(Accident) = 0,05√ó0,60 + 0,15√ó0,30 + 0,40√ó0,10\n   P(Accident) = 0,03 + 0,045 + 0,04 = 0,115",
            "3Ô∏è‚É£ Apliquem Bayes per P(Alt|Accident):\n   P(Alt|Accident) = P(Accident|Alt)√óP(Alt) / P(Accident)\n   P(Alt|Accident) = 0,40√ó0,10 / 0,115 = 0,04 / 0,115 = 0,3478",
            "‚úÖ RESPOSTA: Hi ha un 34,78% de probabilitat que sigui de risc alt"
        ]
    )
]

problemes_prob_total = [
    ProblemaEstadistica(
        enunciat="""
üè≠ PROBLEMA DE QUALITAT

Una f√†brica de tel√®fons m√≤bils t√© 3 l√≠nies de producci√≥:
‚Ä¢ L√≠nia A: Produeix el 45% dels tel√®fons, amb 1% de defectuosos
‚Ä¢ L√≠nia B: Produeix el 35% dels tel√®fons, amb 2% de defectuosos
‚Ä¢ L√≠nia C: Produeix el 20% dels tel√®fons, amb 4% de defectuosos

‚ùì Quina √©s la probabilitat que un tel√®fon escollit a l'atzar sigui defectu√≥s?
        """,
        tipus="probabilitat_total",
        dades={
            'linies': ['A', 'B', 'C'],
            'p_produccio': [0.45, 0.35, 0.20],
            'p_defecte': [0.01, 0.02, 0.04]
        },
        passos_solucio=[
            "1Ô∏è‚É£ Identifiquem les contribucions de cada l√≠nia:\n   ‚Ä¢ L√≠nia A: 45% de producci√≥ √ó 1% defectuosos\n   ‚Ä¢ L√≠nia B: 35% de producci√≥ √ó 2% defectuosos\n   ‚Ä¢ L√≠nia C: 20% de producci√≥ √ó 4% defectuosos",
            "2Ô∏è‚É£ Calculem cada contribuci√≥:\n   ‚Ä¢ Contribuci√≥ A = 0,45 √ó 0,01 = 0,0045\n   ‚Ä¢ Contribuci√≥ B = 0,35 √ó 0,02 = 0,0070\n   ‚Ä¢ Contribuci√≥ C = 0,20 √ó 0,04 = 0,0080",
            "3Ô∏è‚É£ Sumem totes les contribucions:\n   P(Defectu√≥s) = 0,0045 + 0,0070 + 0,0080 = 0,0195",
            "‚úÖ RESPOSTA: Hi ha un 1,95% de probabilitat que sigui defectu√≥s"
        ]
    ),
    ProblemaEstadistica(
        enunciat="""
üöå PROBLEMA DE TRANSPORT

Els estudiants d'un institut arriben de tres maneres:
‚Ä¢ Autob√∫s: 50% dels estudiants, arriben tard el 10% de vegades
‚Ä¢ Cotxe: 30% dels estudiants, arriben tard el 5% de vegades
‚Ä¢ A peu: 20% dels estudiants, arriben tard el 20% de vegades

‚ùì Quina √©s la probabilitat que un estudiant arribi tard?
        """,
        tipus="probabilitat_total",
        dades={
            'linies': ['Autob√∫s', 'Cotxe', 'A peu'],
            'p_produccio': [0.50, 0.30, 0.20],
            'p_defecte': [0.10, 0.05, 0.20]
        },
        passos_solucio=[
            "1Ô∏è‚É£ Identifiquem les contribucions de cada mitj√†:\n   ‚Ä¢ Autob√∫s: 50% estudiants √ó 10% tard\n   ‚Ä¢ Cotxe: 30% estudiants √ó 5% tard\n   ‚Ä¢ A peu: 20% estudiants √ó 20% tard",
            "2Ô∏è‚É£ Calculem cada contribuci√≥:\n   ‚Ä¢ Contribuci√≥ Autob√∫s = 0,50 √ó 0,10 = 0,05\n   ‚Ä¢ Contribuci√≥ Cotxe = 0,30 √ó 0,05 = 0,015\n   ‚Ä¢ Contribuci√≥ A peu = 0,20 √ó 0,20 = 0,04",
            "3Ô∏è‚É£ Sumem totes les contribucions:\n   P(Tard) = 0,05 + 0,015 + 0,04 = 0,105",
            "‚úÖ RESPOSTA: Hi ha un 10,5% de probabilitat d'arribar tard"
        ]
    )
]

problemes_binomial = [
    ProblemaEstadistica(
        enunciat="""
üèÄ PROBLEMA ESPORTIU

Un jugador de b√†squet t√© un 75% d'encert en tirs lliures.
En un partit crucial, li sancionen amb 6 tirs lliures consecutius.

‚ùì Quina √©s la probabilitat que encesti EXACTAMENT 4 tirs?
        """,
        tipus="binomial",
        dades={
            'n': 6,
            'p': 0.75,
            'k_exacte': 4
        },
        passos_solucio=[
            "1Ô∏è‚É£ Identifiquem les dades:\n   ‚Ä¢ n = 6 tirs (intents)\n   ‚Ä¢ p = 0,75 (probabilitat d'encert)\n   ‚Ä¢ k = 4 (√®xits desitjats)",
            "2Ô∏è‚É£ Apliquem la f√≥rmula binomial:\n   P(X=k) = C(n,k) √ó p^k √ó (1-p)^(n-k)\n   P(X=4) = C(6,4) √ó 0,75^4 √ó 0,25^2",
            "3Ô∏è‚É£ Calculem cada part:\n   ‚Ä¢ C(6,4) = 6!/(4!√ó2!) = 15\n   ‚Ä¢ 0,75^4 = 0,3164\n   ‚Ä¢ 0,25^2 = 0,0625\n   ‚Ä¢ P(X=4) = 15 √ó 0,3164 √ó 0,0625 = 0,2966",
            "‚úÖ RESPOSTA: Hi ha un 29,66% de probabilitat d'encertar exactament 4 tirs"
        ]
    ),
    ProblemaEstadistica(
        enunciat="""
üìù PROBLEMA D'EXAMEN

Un estudiant respon 8 preguntes de veritat/fals a l'atzar.
Per aprovar necessita encertar almenys 6 preguntes.

‚ùì Quina √©s la probabilitat que aprovi l'examen?
        """,
        tipus="binomial",
        dades={
            'n': 8,
            'p': 0.5,
            'k_minim': 6
        },
        passos_solucio=[
            "1Ô∏è‚É£ Identifiquem les dades:\n   ‚Ä¢ n = 8 preguntes\n   ‚Ä¢ p = 0,5 (veritat/fals a l'atzar)\n   ‚Ä¢ Necessita k ‚â• 6 per aprovar",
            "2Ô∏è‚É£ Calculem P(X‚â•6) = P(X=6) + P(X=7) + P(X=8):\n   ‚Ä¢ P(X=6) = C(8,6) √ó 0,5^8 = 28 √ó 0,00391 = 0,1094\n   ‚Ä¢ P(X=7) = C(8,7) √ó 0,5^8 = 8 √ó 0,00391 = 0,0313\n   ‚Ä¢ P(X=8) = C(8,8) √ó 0,5^8 = 1 √ó 0,00391 = 0,0039",
            "3Ô∏è‚É£ Sumem les probabilitats:\n   P(Aprovar) = 0,1094 + 0,0313 + 0,0039 = 0,1446",
            "‚úÖ RESPOSTA: Hi ha un 14,46% de probabilitat d'aprovar"
        ]
    )
]

def visualitzar_bayes(dades):
    """Visualitzaci√≥ millorada per a problemes de Bayes amb diagrama d'arbre clar"""
    fig = plt.figure(figsize=(16, 10))

    # Crear subplots
    gs = fig.add_gridspec(2, 2, height_ratios=[3, 1], width_ratios=[1, 1])
    ax_arbre = fig.add_subplot(gs[0, :])
    ax_inicial = fig.add_subplot(gs[1, 0])
    ax_final = fig.add_subplot(gs[1, 1])

    # DIAGRAMA D'ARBRE
    ax_arbre.set_xlim(0, 10)
    ax_arbre.set_ylim(0, 10)
    ax_arbre.axis('off')
    ax_arbre.set_title('DIAGRAMA D\'ARBRE - TEOREMA DE BAYES', fontsize=16, fontweight='bold', pad=20)

    # Obtenir les dades
    if 'p_malaltia' in dades:
        p_condicio = dades['p_malaltia']
        p_test_si = dades['p_positiu_donat_malalt']
        p_test_no = dades['p_positiu_donat_sa']
        nom_condicio = dades.get('nom_condicio', 'Condici√≥')
        nom_test = dades.get('nom_test', 'Test+')
    else:
        # Per problemes amb m√∫ltiples grups
        p_condicio = dades['p_grup'][-1]
        p_test_si = dades['p_accident'][-1]
        p_test_no = sum(dades['p_grup'][i]*dades['p_accident'][i] for i in range(len(dades['p_grup'])-1)) / sum(dades['p_grup'][:-1])
        nom_condicio = dades.get('nom_condicio', 'Condici√≥')
        nom_test = dades.get('nom_test', 'Test+')

    p_no_condicio = 1 - p_condicio
    p_no_test_si = 1 - p_test_si
    p_no_test_no = 1 - p_test_no

    # Dibuixar l'arbre
    # Node inicial
    ax_arbre.add_patch(plt.Circle((2, 5), 0.3, color=COLORS['neutre'], ec='black', linewidth=2))
    ax_arbre.text(2, 5, 'INICI', ha='center', va='center', fontweight='bold')

    # Branques principals
    ax_arbre.plot([2.3, 4], [5, 7], 'k-', linewidth=3)
    ax_arbre.plot([2.3, 4], [5, 3], 'k-', linewidth=3)

    # Nodes de condici√≥
    ax_arbre.add_patch(Rectangle((3.5, 6.5), 2, 1, color=COLORS['vermell_clar'], ec='black', linewidth=2))
    ax_arbre.text(4.5, 7, f'{nom_condicio}\nS√ç', ha='center', va='center', fontweight='bold')
    ax_arbre.text(3, 6, f'{p_condicio:.1%}', ha='center', va='center', fontsize=14, color='red', fontweight='bold')

    ax_arbre.add_patch(Rectangle((3.5, 2.5), 2, 1, color=COLORS['verd_clar'], ec='black', linewidth=2))
    ax_arbre.text(4.5, 3, f'{nom_condicio}\nNO', ha='center', va='center', fontweight='bold')
    ax_arbre.text(3, 4, f'{p_no_condicio:.1%}', ha='center', va='center', fontsize=14, color='green', fontweight='bold')

    # Branques secund√†ries
    ax_arbre.plot([5.5, 7], [7, 8], 'k-', linewidth=2)
    ax_arbre.plot([5.5, 7], [7, 6], 'k-', linewidth=2)
    ax_arbre.plot([5.5, 7], [3, 4], 'k-', linewidth=2)
    ax_arbre.plot([5.5, 7], [3, 2], 'k-', linewidth=2)

    # Nodes finals
    ax_arbre.add_patch(FancyBboxPatch((6.5, 7.7), 2, 0.6, boxstyle="round,pad=0.1",
                                      facecolor=COLORS['groc'], edgecolor='black', linewidth=2))
    ax_arbre.text(7.5, 8, f'{nom_test}', ha='center', va='center', fontsize=11, fontweight='bold')
    ax_arbre.text(6.2, 7.5, f'{p_test_si:.1%}', ha='center', va='center', fontsize=12, color='darkred')

    ax_arbre.add_patch(FancyBboxPatch((6.5, 5.7), 2, 0.6, boxstyle="round,pad=0.1",
                                      facecolor='lightgray', edgecolor='black', linewidth=1))
    ax_arbre.text(7.5, 6, 'Test-', ha='center', va='center', fontsize=11)
    ax_arbre.text(6.2, 6.5, f'{p_no_test_si:.1%}', ha='center', va='center', fontsize=12, color='gray')

    ax_arbre.add_patch(FancyBboxPatch((6.5, 3.7), 2, 0.6, boxstyle="round,pad=0.1",
                                      facecolor=COLORS['groc'], edgecolor='black', linewidth=2))
    ax_arbre.text(7.5, 4, f'{nom_test}', ha='center', va='center', fontsize=11, fontweight='bold')
    ax_arbre.text(6.2, 3.5, f'{p_test_no:.1%}', ha='center', va='center', fontsize=12, color='darkgreen')

    ax_arbre.add_patch(FancyBboxPatch((6.5, 1.7), 2, 0.6, boxstyle="round,pad=0.1",
                                      facecolor='lightgray', edgecolor='black', linewidth=1))
    ax_arbre.text(7.5, 2, 'Test-', ha='center', va='center', fontsize=11)
    ax_arbre.text(6.2, 2.5, f'{p_no_test_no:.1%}', ha='center', va='center', fontsize=12, color='gray')

    # C√†lculs
    p_test_total = p_test_si * p_condicio + p_test_no * p_no_condicio
    p_condicio_donat_test = (p_test_si * p_condicio) / p_test_total

    # Probabilitats conjuntes
    ax_arbre.text(9, 8, f'P({nom_condicio} i {nom_test}) = {p_condicio*p_test_si:.4f}',
                  fontsize=11, bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))
    ax_arbre.text(9, 4, f'P(No {nom_condicio} i {nom_test}) = {p_no_condicio*p_test_no:.4f}',
                  fontsize=11, bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue", alpha=0.7))

    # GR√ÄFIC PROBABILITAT INICIAL
    ax_inicial.bar([nom_condicio], [p_condicio], color=COLORS['secundari'], alpha=0.7)
    ax_inicial.set_ylim(0, max(0.1, p_condicio * 1.5))
    ax_inicial.set_title(f'Probabilitat INICIAL\nP({nom_condicio}) = {p_condicio:.1%}', fontsize=14, fontweight='bold')
    ax_inicial.set_ylabel('Probabilitat')

    # GR√ÄFIC PROBABILITAT ACTUALITZADA
    ax_final.bar([f'{nom_condicio}|{nom_test}'], [p_condicio_donat_test], color=COLORS['principal'])
    ax_final.set_ylim(0, max(0.5, p_condicio_donat_test * 1.2))
    ax_final.set_title(f'Probabilitat ACTUALITZADA\nP({nom_condicio}|{nom_test}) = {p_condicio_donat_test:.1%}',
                       fontsize=14, fontweight='bold')
    ax_final.set_ylabel('Probabilitat')

    # Afegir la f√≥rmula de Bayes
    formula_text = f"P({nom_condicio}|{nom_test}) = P({nom_test}|{nom_condicio}) √ó P({nom_condicio}) / P({nom_test})"
    fig.text(0.5, 0.02, formula_text, ha='center', fontsize=14,
             bbox=dict(boxstyle="round,pad=0.5", facecolor="lightyellow", edgecolor="black"))

    plt.tight_layout()
    plt.show()

def visualitzar_probabilitat_total(dades):
    """Visualitzaci√≥ millorada per a probabilitat total"""
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))

    # Dades
    noms = dades['linies']
    probabilitats = dades['p_produccio']
    prob_esdeveniment = dades['p_defecte']

    # 1. Diagrama de sectors
    colors = [COLORS['principal'], COLORS['secundari'], COLORS['groc']]
    wedges, texts, autotexts = ax1.pie(probabilitats, labels=noms, colors=colors, autopct='%1.1f%%',
                                       startangle=90, textprops={'fontsize': 12, 'fontweight': 'bold'})
    ax1.set_title('Distribuci√≥ per Origen', fontsize=14, fontweight='bold')

    # 2. Contribucions individuals
    contribucions = [p * e for p, e in zip(probabilitats, prob_esdeveniment)]
    bars = ax2.bar(noms, contribucions, color=colors, alpha=0.7, edgecolor='black', linewidth=2)

    for bar, contrib, prob, event in zip(bars, contribucions, probabilitats, prob_esdeveniment):
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height + 0.001,
                f'{contrib:.4f}\n({event:.1%}√ó{prob:.1%})',
                ha='center', va='bottom', fontsize=10, fontweight='bold')

    ax2.set_ylabel('Contribuci√≥ a la Probabilitat Total', fontsize=12)
    ax2.set_title('Contribuci√≥ de Cada Origen', fontsize=14, fontweight='bold')
    ax2.set_ylim(0, max(contribucions) * 1.3)

    # 3. Suma total visualitzada
    total = sum(contribucions)
    ax3.bar(['TOTAL'], [total], color=COLORS['exit'], edgecolor='black', linewidth=3)
    ax3.text(0, total/2, f'{total:.4f}\n({total*100:.2f}%)',
             ha='center', va='center', fontsize=16, fontweight='bold', color='white')

    y_pos = total * 1.1
    for i, (nom, contrib) in enumerate(zip(noms, contribucions)):
        ax3.text(0.3, y_pos - i*0.02, f'{nom}: {contrib:.4f}',
                fontsize=11, ha='left')
        if i < len(contribucions) - 1:
            ax3.text(0.25, y_pos - i*0.02 - 0.01, '+', fontsize=12, ha='left')

    ax3.axhline(y=y_pos - len(contribucions)*0.02, color='black', linewidth=1)
    ax3.text(0.25, y_pos - len(contribucions)*0.02 - 0.02, f'= {total:.4f}',
            fontsize=12, fontweight='bold', ha='left')

    ax3.set_ylabel('Probabilitat Total', fontsize=12)
    ax3.set_title('Probabilitat Total = Œ£ Contribucions', fontsize=14, fontweight='bold')
    ax3.set_ylim(0, total * 1.4)
    ax3.set_xlim(-0.5, 0.8)

    plt.tight_layout()
    plt.show()

def visualitzar_binomial(dades):
    """Visualitzaci√≥ millorada per a distribuci√≥ binomial"""
    n = dades['n']
    p = dades['p']

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

    # 1. Distribuci√≥ de probabilitat
    x = np.arange(0, n+1)
    probs = [stats.binom.pmf(k, n, p) for k in x]

    bars = ax1.bar(x, probs, color=COLORS['neutre'], edgecolor='black', alpha=0.5, linewidth=2)

    if 'k_exacte' in dades:
        k = dades['k_exacte']
        bars[k].set_color(COLORS['principal'])
        bars[k].set_alpha(1.0)
        bars[k].set_linewidth(3)
        ax1.axvline(x=k, color='red', linestyle='--', alpha=0.7, linewidth=2)
        ax1.text(k, max(probs)*1.1, f'k = {k}', ha='center', fontsize=14,
                bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))

    if 'k_minim' in dades:
        k_min = dades['k_minim']
        for k in range(k_min, n+1):
            bars[k].set_color(COLORS['exit'])
            bars[k].set_alpha(1.0)
            bars[k].set_linewidth(2)

        ax1.axvspan(k_min-0.5, n+0.5, alpha=0.2, color=COLORS['exit'])
        ax1.text((k_min+n)/2, max(probs)*1.15, f'k ‚â• {k_min}', ha='center', fontsize=14,
                bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen", alpha=0.7))

    for i, (xi, prob) in enumerate(zip(x, probs)):
        if prob > 0.01:
            ax1.text(xi, prob + 0.005, f'{prob:.3f}', ha='center', va='bottom', fontsize=9)

    ax1.set_xlabel('Nombre d\'√®xits (k)', fontsize=12)
    ax1.set_ylabel('Probabilitat P(X=k)', fontsize=12)
    ax1.set_title(f'Distribuci√≥ Binomial B({n}, {p})', fontsize=14, fontweight='bold')
    ax1.set_xticks(x)
    ax1.grid(True, alpha=0.3, axis='y')
    ax1.set_ylim(0, max(probs) * 1.3)

    # 2. C√†lcul visual
    ax2.axis('off')
    ax2.set_xlim(0, 10)
    ax2.set_ylim(0, 10)
    ax2.text(5, 9, 'C√ÄLCUL DE LA PROBABILITAT', ha='center', fontsize=16, fontweight='bold')

    y_pos = 7.5

    if 'k_exacte' in dades:
        k = dades['k_exacte']
        prob = stats.binom.pmf(k, n, p)

        ax2.text(5, y_pos, f'P(X = {k}) = C({n},{k}) √ó {p}^{k} √ó {1-p}^{n-k}',
                ha='center', fontsize=14, bbox=dict(boxstyle="round,pad=0.3", facecolor="lightyellow"))

        comb = int(stats.binom.pmf(k, n, p) / (p**k * (1-p)**(n-k)))
        ax2.text(5, y_pos-1.5, f'C({n},{k}) = {comb}', ha='center', fontsize=12)
        ax2.text(5, y_pos-2.5, f'{p}^{k} = {p**k:.4f}', ha='center', fontsize=12)
        ax2.text(5, y_pos-3.5, f'{1-p}^{n-k} = {(1-p)**(n-k):.4f}', ha='center', fontsize=12)
        ax2.text(5, y_pos-5, f'= {comb} √ó {p**k:.4f} √ó {(1-p)**(n-k):.4f}', ha='center', fontsize=12)
        ax2.text(5, y_pos-6.5, f'= {prob:.4f} ({prob*100:.2f}%)', ha='center', fontsize=16,
                fontweight='bold', bbox=dict(boxstyle="round,pad=0.5", facecolor="lightgreen"))

    elif 'k_minim' in dades:
        k_min = dades['k_minim']
        probs_sum = [stats.binom.pmf(k, n, p) for k in range(k_min, n+1)]

        ax2.text(5, y_pos, f'P(X ‚â• {k_min}) = ', ha='center', fontsize=14)

        for i, k in enumerate(range(k_min, n+1)):
            prob_k = stats.binom.pmf(k, n, p)
            ax2.text(5, y_pos-1.5-i*0.8, f'P(X={k}) = {prob_k:.4f}', ha='center', fontsize=11)
            if k < n:
                ax2.text(5, y_pos-1.9-i*0.8, '+', ha='center', fontsize=11)

        total = sum(probs_sum)
        ax2.text(5, y_pos-2-len(probs_sum)*0.8, '‚Äï‚Äï‚Äï‚Äï‚Äï‚Äï‚Äï', ha='center', fontsize=11)
        ax2.text(5, y_pos-3-len(probs_sum)*0.8, f'= {total:.4f} ({total*100:.2f}%)',
                ha='center', fontsize=16, fontweight='bold',
                bbox=dict(boxstyle="round,pad=0.5", facecolor="lightgreen"))

    plt.tight_layout()
    plt.show()

def mostrar_problema_amb_solucio(problema):
    """Mostra un problema i la seva soluci√≥ pas a pas"""
    print("\n" + "="*70)
    print("üìö PROBLEMA")
    print("="*70)
    print(problema.enunciat)

    input("\nPrem ENTER per veure la soluci√≥ pas a pas...")

    print("\n" + "="*70)
    print("üìä SOLUCI√ì PAS A PAS")
    print("="*70)

    for pas in problema.passos_solucio:
        print(f"\n{pas}")
        time.sleep(1)

    print("\n\nüìà Generant visualitzaci√≥...")
    time.sleep(1)

    # Visualitzar segons el tipus
    if problema.tipus == 'bayes':
        visualitzar_bayes(problema.dades)
    elif problema.tipus == 'probabilitat_total':
        visualitzar_probabilitat_total(problema.dades)
    elif problema.tipus == 'binomial':
        visualitzar_binomial(problema.dades)

def menu_principal():
    """Men√∫ principal del programa"""
    while True:
        print("\n" + "="*70)
        print("üéì FITXA D'ESTAD√çSTICA - MATEM√ÄTIQUES BATXILLERAT 2")
        print("="*70)
        print("\nTRIA EL TIPUS DE PROBLEMA QUE VOLS ESTUDIAR:")
        print("\n1Ô∏è‚É£  TEOREMA DE BAYES")
        print("    ‚Üí Actualitzar probabilitats amb nova informaci√≥")
        print("\n2Ô∏è‚É£  PROBABILITAT TOTAL")
        print("    ‚Üí Sumar probabilitats de diferents camins")
        print("\n3Ô∏è‚É£  DISTRIBUCI√ì BINOMIAL")
        print("    ‚Üí Comptar √®xits en experiments repetits")
        print("\n4Ô∏è‚É£  SORTIR")
        print("\n" + "-"*70)

        opcio = input("\nTria una opci√≥ (1-4): ")

        if opcio == '1':
            problema = random.choice(problemes_bayes)
            mostrar_problema_amb_solucio(problema)
            continuar = input("\n\nüîÑ Vols veure un altre problema de Bayes? (s/n): ")
            if continuar.lower() == 's':
                problema = random.choice(problemes_bayes)
                mostrar_problema_amb_solucio(problema)

        elif opcio == '2':
            problema = random.choice(problemes_prob_total)
            mostrar_problema_amb_solucio(problema)
            continuar = input("\n\nüîÑ Vols veure un altre problema de Probabilitat Total? (s/n): ")
            if continuar.lower() == 's':
                problema = random.choice(problemes_prob_total)
                mostrar_problema_amb_solucio(problema)

        elif opcio == '3':
            problema = random.choice(problemes_binomial)
            mostrar_problema_amb_solucio(problema)
            continuar = input("\n\nüîÑ Vols veure un altre problema Binomial? (s/n): ")
            if continuar.lower() == 's':
                problema = random.choice(problemes_binomial)
                mostrar_problema_amb_solucio(problema)

        elif opcio == '4':
            print("\n\nüéì Gr√†cies per utilitzar la fitxa d'estad√≠stica!")
            print("üí™ Molta sort en els teus estudis!")
            break
        else:
            print("\n‚ùå Opci√≥ no v√†lida. Tria 1, 2, 3 o 4")

# EXECUTAR DIRECTAMENT
if __name__ == "__main__":
    menu_principal()
```
